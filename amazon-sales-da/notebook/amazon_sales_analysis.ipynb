## 1. Data loading

```python
import pandas as pd
df = pd.read_csv("Amazon Sale Report.csv")
df.head()
```

```python
df.info()
df.describe()
```

## 2. Data Cleaning

```python
df=df.drop(columns=["Unnamed: 22"],errors="ignore")
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%m-%d-%y")
df["Month"] = df["Date"].dt.to_period("M")
```

## 3. Data Analysis

```python
df.groupby("Category")["Amount"].sum().sort_values(ascending=False)
```

```python
df.groupby("ship-state")["Amount"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby("Month")["Amount"].sum()
```

```python
df["Status"].value_counts()
```

```python
cancel_rate = (df["Status"] == "Cancelled").mean()
```

```python
import matplotlib.pyplot as plt
```

```python
monthly_sales=df.groupby("Month")["Amount"].sum()
```

```python
df.groupby("B2B")["Amount"].sum()
```

```python
df.groupby("ship-city")["Amount"].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby("Fulfilment")["Amount"].sum()
```

```python
df["Weekday"] = df["Date"].dt.day_name()
df.groupby("Weekday")["Amount"].sum()
```

## 4. Visualizations

```python
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.savefig("images/revenue_trend.png")
```

```python
weekdays=df.groupby("Weekday")["Amount"].sum()
weekdays.plot()
plt.xlabel("Weekdays")
plt.ylabel("Objednavky")
```

```python
state_sales=df.groupby("ship-state")["Amount"].sum().sort_values(ascending=False).head(10)

state_sales.plot(kind="bar")
plt.title("Top 10 States by Revenue")
plt.xlabel("State")
plt.ylabel("Revenue")
plt.xticks(rotation=45)

plt.show()
```

```python
monthly_sales = df.groupby("Month")["Amount"].sum()

monthly_sales.plot()

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

```python
status_counts = df["Status"].value_counts()

status_counts.plot(kind="bar")

plt.title("Order Status Distribution")
plt.xlabel("Status")
plt.ylabel("Number of Orders")

plt.show()
```

```python
category_sales = df.groupby("Category")["Amount"].sum().sort_values(ascending=False)

category_sales.plot(kind="bar")

plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.show()
```

## 5. Business Insights

Key Business Insights

- The majority of revenue is generated by the "Set" category, indicating strong customer demand for bundled products.
- The state of Maharashtra and the city of Bengaluru generate the highest revenue, suggesting these regions are key markets.
- Sales peaked in March 2022, indicating possible seasonal demand patterns.
- The cancellation rate appears relatively high, which could indicate potential issues in fulfilment or customer expectations.
- Order distribution by weekday shows the highest activity on Sundays and the lowest on Thursdays.
